```python
import numpy as np
import pandas as pd

#kmeans
from sklearn.cluster import KMeans

#plot stuff
import matplotlib.pyplot as mpl
from mpl_toolkits.mplot3d import Axes3D as plt
```

```python
#tsne
from sklearn.manifold import TSNE

#make sample dataset. should've reversed dimensions, but can just .transpose() later
arr = np.array([[0,0,5,  40,45, 40,      0,0,  0,   40,40,35,35],
                [0,5,0,  30,  30, 30,    30,30,25,  5,0,  0, 5],
                [0,0,0,  0, 5,5,         25,30,30,  0,0,  0,  0]])

#run tsne for 3 dimensions
tsne = TSNE(n_components = 3,n_iter = 5000).fit_transform(arr.transpose())
print(tsne.shape)

#plot
mpl.figure(0)
ax = mpl.axes(projection = '3d')
ax.scatter(arr[0],arr[1],arr[2])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

mpl.figure(1)
ax = mpl.axes(projection = '3d')
ax.scatter(tsne[:,0],tsne[:,1],tsne[:,2])
mpl.show()
```

```python
#pca 
from sklearn.decomposition import PCA

#run pca for 2 dimensions
pca = PCA(n_components = 2).fit_transform(arr.transpose())

#plot
mpl.figure(3)
mpl.scatter(pca[:,0],pca[:,1])
mpl.show()
```

```python
#run kmeans on PCA results
kmeans = KMeans(n_clusters = 4)
kmeans.fit(pca)

predict = kmeans.predict(pca)

#plot
mpl.scatter(pca[:, 0], pca[:, 1], c=predict)
```
